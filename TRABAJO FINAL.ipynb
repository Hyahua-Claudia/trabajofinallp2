```python
sopa = BeautifulSoup(resp.text, "html.parser")
```

```python
reseñas = sopa.find_all("div", {"class": "a-section review aok-relative"})
reseñas
```

```python
#funciones
```

```python
def obtener_reseña_estrellas(soup_obj: BeautifulSoup) -> str:
    try:
        review_str = soup_obj.find('a', {"class": "a-link-normal"}).get_text()
        return review_str
    except Exception as e:
        return 'no_stars'
        print(e)
```

```python
def obtener_reseña_cuerpo(soup_obj: BeautifulSoup) -> str:
    try:
        review_body = soup_obj.find('span', 
            {"class": "a-size-base review-text review-text-content"}
        ).get_text().strip()
        return review_body
    except Exception as e:
        return 'no_body'
        print(e)
```

```python
def obtener_reseña_fecha(soup_obj: BeautifulSoup) -> str:
    try:
        review_date = soup_obj.find('span', 
            {"class": "a-size-base a-color-secondary review-date"}
        ).get_text().strip()
        return review_date
    except Exception as e:
        return 'no_date'
        print(e)
```

```python
def obtener_reseña_componentes(soup_obj: BeautifulSoup) -> dict:
    return {
        'ESTRELLAS': obtener_reseña_estrellas(soup_obj) , 
        'Comentario': obtener_reseña_cuerpo(soup_obj), 
        'FECHA': obtener_reseña_fecha(soup_obj)
    }
```

```python
records = [obtener_reseña_componentes(rev) for rev in reseñas]
DF = pd.DataFrame.from_records(records)
DF
```
